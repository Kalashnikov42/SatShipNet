```python
import torch
import os
from pathlib import Path
from PIL import Image
import numpy as np
from ultralytics import YOLO


train_image_dir = Path(r'C:\Users\kalas\OneDrive\Desktop\SHIPSIZEWORK\train\images')
test_image_dir = Path(r'C:\Users\kalas\OneDrive\Desktop\SHIPSIZEWORK\test\images')


model = YOLO('yolov8n.pt')  


SHIP_CLASS_ID = 8 


def extract_bounding_boxes(image_path):
    img = Image.open(image_path)
    results = model(img)  
    bounding_boxes = []
    for result in results: 
        for box in result.boxes:  
            cls = int(box.cls[0].item())  
            if cls == SHIP_CLASS_ID:  
                x_min, y_min, x_max, y_max = box.xyxy[0].tolist()[:4] 
                confidence = box.conf[0].item() 
                width = x_max - x_min
                height = y_max - y_min
                bounding_boxes.append((x_min, y_min, width, height, confidence, cls))
    return bounding_boxes


def process_image_directory(image_dir):
    all_bounding_boxes = {}
    for image_file in image_dir.glob('*.*'):
        bounding_boxes = extract_bounding_boxes(image_file)
        all_bounding_boxes[image_file.name] = bounding_boxes
    return all_bounding_boxes


train_bounding_boxes = process_image_directory(train_image_dir)

for image_name, boxes in train_bounding_boxes.items():
 #   print(f"Training Image: {image_name}")
    for box in boxes:
        print(f"Bounding Box: {box}")


test_bounding_boxes = process_image_directory(test_image_dir)

for image_name, boxes in test_bounding_boxes.items():
 #   print(f"Testing Image: {image_name}")
    for box in boxes:
        print(f"Bounding Box: {box}")

with open('bounding_boxes.txt', 'w') as f:
    f.write("Training Dataset:\n")
    for image_name, boxes in train_bounding_boxes.items():
     #   f.write(f"Image: {image_name}\n")
        for box in boxes:
            f.write(f"Bounding Box: {box}\n")
    f.write("Testing Dataset:\n")
    for image_name, boxes in test_bounding_boxes.items():
      #  f.write(f"Image: {image_name}\n")
        for box in boxes:
            f.write(f"Bounding Box: {box}\n")
```
